```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

# Parameters for the grid
rows = 6
cols = 16
circle_radius = 0.3
circle_col_spacing = circle_radius*2
circle_row_spacing = circle_radius*3.8  # Space between circles
rectangle_width = circle_col_spacing*0.42
rectangle_height = circle_radius*0.16

# Create a figure
fig, ax = plt.subplots(figsize=((cols+1)*circle_col_spacing, (rows+1)*circle_row_spacing))

# Set the aspect of the plot to be equal
ax.set_aspect('equal')

# Loop through grid to plot circles
for row in range(rows):
    for col in range(cols):
        slide=6;
        # Alternate colors
        if (col) % 2 == 0:
            color = "blue"
            rect_color1="blue"
            rect_color2="yellow"
        else:
            color = "yellow"
            rect_color2="blue"
            rect_color1="yellow"
        if(row%2):
          temp=rect_color1
          rect_color1=rect_color2
          rect_color2=temp
          slide*=-1;


        if col!=cols-1:
          rect1=patches.Rectangle(((col+0.5) * circle_col_spacing-rectangle_width/2,
                                       -row * circle_row_spacing +circle_radius - rectangle_height- (slide/400)),
                                      rectangle_width, rectangle_height,
                                      angle=slide, color =rect_color1)
          rect2=rect_right = patches.Rectangle(((col+0.5) * circle_col_spacing-rectangle_width/2,
                                        -row * circle_row_spacing -circle_radius + rectangle_height/2 + (abs(slide)*-1/400)),
                                       rectangle_width, rectangle_height,
                                       angle=slide, color = rect_color2)
          ax.add_artist(rect1)
          ax.add_artist(rect2)

        # Position for each circle
        circle = plt.Circle((col * circle_col_spacing, -row * circle_row_spacing),
                            circle_radius, color=color)
        ax.add_artist(circle)


# Set limits and background color
ax.set_xlim(-1, cols * circle_col_spacing)
ax.set_ylim(-rows * circle_row_spacing, 1)
ax.set_facecolor("gray")

# Remove axes for cleaner look
#ax.axis('off')

# Display the plot
plt.show()

```
